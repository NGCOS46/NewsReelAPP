```python
for i, url in enumerate(urls):
    # Send HTTP request
    result = req.get(url)

    # Check if the request was successful
    if result.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(result.text, "html.parser")

        # Extract information from the HTML
        articles = soup.find_all('div', class_='CONTROL_ITEM item-hitlist mosaico')

        # Iterate through the articles
        for article in articles:
            article_info = {
                'date': article.find('p', class_='fecha').text.strip(),
                'description': article.find('p', class_='titulo text-bold').text.strip(),
                'genero': i + 1  # Assign a 'genero' value based on the index of the URL in the list
            }
            articles_list.append(article_info)           
```

```python
df= pd.DataFrame(articles_list)
df
```

```python
df['time'] = df['date'].str.extract(r'\b(\d{2}:\d{2})\b')

# If there's a possibility that 'pubDate' contains null values, you can fill them with '00:00'
df['time'] = df['time'].fillna('00:00')
df
```

```python
df['date'] = df['date'].str.extract(r'(\d{2}/\d{2}/\d{4})')
df['date']= pd.to_datetime(df['date'], format='%d/%m/%Y')

df
```
